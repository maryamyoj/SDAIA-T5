## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('C:\\Users\\DELL\\Downloads\\Task 2-20240725T044843Z-001\\Task 2\\Task (2) Dataset.csv')
df
```

### Identify and analyze patterns in the data.

```python
numerics= df.select_dtypes(include=['float', 'int'])
corrMat = numerics.corr()
sns.heatmap(corrMat, annot=True)
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
df.plot.hexbin(x='JamsCount', y='JamsDelay', gridsize=15)
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
df.plot.scatter(x='City', y='JamsCount')
```
